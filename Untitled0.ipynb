## 1. Мета роботи
Мета роботи — навчитись виявляти обличчя, очі, посмішки, а також пішоходів з використанням каскадів Хаара та HOG-класифікаторів бібліотеки OpenCV.

## 2. Програма роботи
Встановлення необхідних бібліотек та завантаження ресурсів.

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np

# Завантаження каскадів
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml')
```

## 3. Виявлення облич, очей та посмішки на груповому фото
Згідно з завданням, використовуємо фото з декількома людьми для підрахунку їх кількості та виявлення деталей обличчя.

```python

img = cv2.imread('/content/drive/MyDrive/Lab2/people.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, 1.1, 3)

count_people = len(faces)
cv2.putText(img, f'Total People: {count_people}', (20, 50), cv2.FONT_HERSHEY_DUPLEX, 1.0, (0, 0, 0), 2)

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)

    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]

    eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 10)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)

    smiles = smile_cascade.detectMultiScale(roi_gray, 1.3, 15)
    for (sx, sy, sw, sh) in smiles:
        if sy > h/2:
            cv2.rectangle(roi_color, (sx, sy), (sx+sw, sy+sh), (0, 0, 255), 2)

cv2_imshow(img)
```

## 4. Розпізнавання на відео
Реалізація виявлення облич на відео тривалістю не менше 30 секунд.

```python
video_path = '/content/drive/MyDrive/Lab2/norm.mp4'
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Помилка: Не вдалося відкрити відео файл!")
else:
    fps = cap.get(cv2.CAP_PROP_FPS) or 25.0
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    output_name = '/content/drive/MyDrive/Lab2/faces_result_large.mp4'
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_name, fourcc, fps, (width, height))

    print(f"Початок обробки... (Роздільна здатність: {width}x{height})")

    frame_count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        gray_frame = cv2.equalizeHist(gray_frame)

        faces = face_cascade.detectMultiScale(
            gray_frame,
            scaleFactor=1.1,
            minNeighbors=3,
            minSize=(180, 180)
        )

        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
            cv2.putText(frame, 'Face', (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

        out.write(frame)

        frame_count += 1
        if frame_count % 50 == 0:
            print(f"Оброблено кадрів: {frame_count}")

    cap.release()
    out.release()
    print(f"Готово! Результат збережено у файл: {output_name}")
```

## 5. Виявлення пішоходів за допомогою HOG-класифікатора
Обробимо відео для виділення пішоходів.

```python

hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

video_path = '/content/drive/MyDrive/Lab2/vidos piplov.mp4'
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Помилка: Не вдалося відкрити відео файл!")
else:
    fps = cap.get(cv2.CAP_PROP_FPS)
    if fps == 0 or fps is None: fps = 24.0

    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    output_name = '/content/drive/MyDrive/Lab2/pedestrians_result.mp4'
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_name, fourcc, fps, (width, height))

    print(f"Починаємо пошук пішоходів... (Роздільна здатність: {width}x{height})")

    frame_count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret: break

        img_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        (rects, weights) = hog.detectMultiScale(
            img_gray,
            winStride=(8, 8),
            padding=(8, 8),
            scale=1.1
        )

        for (x, y, w, h) in rects:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(frame, 'Person', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        out.write(frame)

        frame_count += 1
        if frame_count % 50 == 0:
            print(f"Оброблено кадрів: {frame_count}")

    cap.release()
    out.release()
    print(f"Готово! Результат збережено у файл: {output_name}")
```

### Висновок
У ході лабораторної роботи було реалізовано алгоритми детекції об'єктів за допомогою OpenCV:
* **Каскади Хаара:** успішно застосовані для пошуку облич та очей. Встановлено, що параметри `scaleFactor` та `minNeighbors` є критичними для відсікання хибних спрацювань (наприклад, неба чи елементів фону).
* **Лічильник:** додано функцію автоматичного підрахунку виявлених осіб на зображенні.
* **HOG-дескриптор:** використаний для детекції пішоходів. Метод точний для розпізнавання силуетів у повний зріст, проте потребує значних обчислювальних ресурсів порівняно з каскадами Хаара.
* **Оптимізація:** налаштовано параметри `winStride` та `scale` для прискорення обробки відео без суттєвої втрати точності.

## Контрольні запитання
1. **Алгоритм Віоли-Джонса** — це метод виявлення об’єктів на зображенні в реальному часі, що базується на ознаках Хаара та каскадних класифікаторах.
2. **Haarcascade** — це XML-файл, що містить результати навчання класифікатора для виявлення специфічних об'єктів (облич, очей тощо).
3. **HOG-класифікатор** (Histogram of Oriented Gradients) — дескриптор ознак, що використовується для виявлення об'єктів шляхом підрахунку орієнтацій градієнта.
4. **cvtColor** — метод для перетворення колірного простору зображення (наприклад, з BGR у Grayscale для прискорення обробки).
